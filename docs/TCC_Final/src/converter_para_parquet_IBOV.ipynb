```python
import pandas as pd
import os
import pyarrow

def analisar_e_converter():
    """
    Função que lê um arquivo CSV financeiro, exibe uma análise básica 
    e salva os dados no formato Parquet para garantir a integridade.
    """
    # Nomes dos arquivos de entrada e saída
    diretorio_dados = r"C:\VSCodeWorkspace\TCC_Final\data"
    arquivo_entrada = 'IBOV_2010_2026.xlsx'
    arquivo_saida = 'IBOV_2010_2026.parquet'
    
    print("--------------------------------------------------")
    print(f"Passo 1: Iniciando a leitura do arquivo {arquivo_entrada}...")
    
    try:
        # Lendo o CSV
        # sep=';' -> Indica que as colunas são separadas por ponto e vírgula
        # decimal=',' -> Garante que os valores financeiros com vírgula sejam lidos como números
        df = pd.read_excel(os.path.join(diretorio_dados, arquivo_entrada))
        print("Leitura concluída com sucesso!\n")
        
        print("--------------------------------------------------")
        print("Passo 2: Análise Rápida dos Dados")
        
        # Mostra a quantidade de linhas e colunas (formato da tabela)
        linhas, colunas = df.shape
        print(f"A base possui {linhas} linhas e {colunas} colunas.\n")
        
        # Exibe as informações das colunas e os tipos de dados
        print("Informações estruturais da base:")
        df.info()
        
        # Mostra as 5 primeiras linhas para conferência visual
        print("\nVisualizando as primeiras 5 linhas:")
        print(df.head())
        print("\n--------------------------------------------------")
        
        print("Passo 3: Salvando no formato Parquet...")
        
        # Salvando a base. index=False evita que o número da linha vire uma coluna inútil
        # engine='pyarrow' é o motor padrão e seguro para gerar o parquet
        caminho_saida = os.path.join(diretorio_dados, arquivo_saida)
        df.to_parquet(caminho_saida, index=False, engine='pyarrow')
        
        print(f"Sucesso! Seus dados foram consolidados e salvos como: {caminho_saida}")
        print("A integridade dos seus dados está garantida no novo formato.")
        
    except FileNotFoundError:
        # Caso o Python não ache o arquivo, damos uma mensagem amigável
        print(f"Erro: Não consegui encontrar o arquivo '{arquivo_entrada}'.")
        print("Verifique se ele está salvo na mesma pasta onde este script está rodando.")
    except Exception as e:
        # Captura qualquer outro erro inesperado
        print(f"Ops, ocorreu um erro inesperado: {e}")

# Este comando faz com que a função rode quando executarmos o script
if __name__ == "__main__":
    analisar_e_converter()
```
