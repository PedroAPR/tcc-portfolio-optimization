Célula 1: Importação das bibliotecas e configuração

```python
import pandas as pd
import os

# Definindo os caminhos dos arquivos

ARQUIVO_ENTRADA = r"C:\VSCodeWorkspace\bkp\COTAHIST_Consolidado_2010_2026.csv"
DIRETORIO_SAIDA = r"C:\VSCodeWorkspace\TCC_Final\data\Ibov"
ARQUIVO_SAIDA = os.path.join(DIRETORIO_SAIDA, "IBOV_Diario_2010_2026.csv")

print("Configurações iniciais carregadas com sucesso!")
```

Célula 2: Leitura, filtragem e exportação dos dados
Nota: Como arquivos consolidados de 2010 a 2026 podem ser muito grandes, o código abaixo faz a leitura eficiente (tratando os dados como texto inicialmente para evitar perda de zeros à esquerda) e realiza a filtragem do ticker do Ibovespa.

```python
try:
    print("Lendo o arquivo consolidado...")
    
    # Lendo o arquivo usando o separador correto ';' do seu ambiente
    df = pd.read_csv(
        ARQUIVO_ENTRADA, 
        sep=';', 
        dtype=str, 
        on_bad_lines='skip'
    )
    
    print(f"Arquivo carregado! Total de linhas: {len(df)}")
    
    # Remove espaços ocultos das colunas
    df.columns = df.columns.str.strip()
    
    coluna_ticker = 'CODIGO_DE_NEGOCIACAO'
    coluna_data = 'DATA_DO_PREGAO'
    
    # Limpa os espaços em branco nos dados da coluna de tickers
    df[coluna_ticker] = df[coluna_ticker].str.strip()
    
    print("Extraindo todas as séries e derivativos vinculados ao Ibovespa ('IBOV')...")
    # Filtra qualquer ticker que comece com 'IBOV' (ex: IBOVC170, IBOVB130, IBOV)
    df_ibov = df[df[coluna_ticker].str.startswith('IBOV', na=False)].copy()
    
    if len(df_ibov) > 0:
        # Ordena cronologicamente pela data do pregão
        df_ibov = df_ibov.sort_values(by=coluna_data)
        
        # Salva o arquivo final tratado
        df_ibov.to_csv(ARQUIVO_SAIDA, index=False, sep=';', encoding='utf-8')
        print("-" * 50)
        print(f"Sucesso! O arquivo do Ibovespa foi gerado com {len(df_ibov)} linhas.")
        print(f"Salvo em: {ARQUIVO_SAIDA}")
    else:
        print("-" * 50)
        print("Erro: Nenhum dado que começa com 'IBOV' foi encontrado no arquivo de entrada.")

except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```
