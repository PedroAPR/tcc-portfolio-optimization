```python
import pandas as pd
import plotly.express as px
import os

def visualizar_evolucao_ativos(diretorio_dados, lista_ativos=None, top_n=136):
    print("=== VISUALIZADOR DE EVOLUÇÃO DE AÇÕES Mínima Variância ===")
    
    # Carregar o painel que acabámos de criar
    caminho_painel = os.path.join(diretorio_dados, "painel_alocacao_mensal_minima_variancia.csv")
    df_painel = pd.read_csv(caminho_painel, parse_dates=['Data'])
    
    # Se o utilizador não pedir ativos específicos, vamos descobrir quais foram os ativos
    # que, em média, tiveram mais peso na carteira ao longo de todos os anos.
    if lista_ativos is None:
        maiores_ativos = df_painel.groupby('Ativo')['Peso'].mean().sort_values(ascending=False).head(top_n).index.tolist()
        df_filtrado = df_painel[df_painel['Ativo'].isin(maiores_ativos)]
        titulo = f"Evolução Dinâmica de Alocação: Top {top_n} Maiores Posições Históricas (Mínima Variância)"
    else:
        df_filtrado = df_painel[df_painel['Ativo'].isin(lista_ativos)]
        titulo = f"Evolução Dinâmica de Alocação: Ativos Selecionados (Mínima Variância) "
        
    # Desenhar o gráfico de linhas interativo
    fig = px.line(df_filtrado, x='Data', y='Peso', color='Ativo', 
                  title=titulo,
                  labels={'Peso': 'Peso na Carteira (%)', 'Data': 'Ano'},
                  line_shape='spline') # spline deixa a linha mais suave
    
    fig.update_layout(yaxis_tickformat='.1%', template='plotly_white')
    fig.show()

# ==========================================
# ÁREA DE EXECUÇÃO
# ==========================================
pasta_dados = r"C:\VSCodeWorkspace\TCC_Escrito\data"

# Para veres os 7 ativos mais importantes do teu modelo:
visualizar_evolucao_ativos(pasta_dados, top_n=136)

# Para veres a "batalha" entre duas ações específicas (exemplo):
# visualizar_evolucao_ativos(pasta_dados, lista_ativos=['VALE3', 'ITUB4'])
```
