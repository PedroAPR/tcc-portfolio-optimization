```python
import pandas as pd
import numpy as np
import os

def processar_fatores_nefin_local(diretorio_dados, nome_arquivo="nefin_factors.csv"):
    print("=== PROCESSAMENTO DE FATORES DE RISCO (FAMA-FRENCH/NEFIN) ===")
    
    caminho_arquivo = os.path.join(diretorio_dados, nome_arquivo)
    
    try:
        # 1. Ler o arquivo CSV local
        print(f"A ler o arquivo local: {nome_arquivo}...")
        df_fatores = pd.read_csv(caminho_arquivo)
        
        # 2. Formatação do Índice de Tempo (Data)
        df_fatores['Date'] = pd.to_datetime(df_fatores['Date'])
        df_fatores.set_index('Date', inplace=True)
        
        # 3. Selecionar apenas as colunas úteis e descartar colunas "lixo" (Unnamed)
        colunas_fatores = ['Rm_minus_Rf', 'SMB', 'HML', 'WML', 'IML']
        df_fatores = df_fatores[colunas_fatores]
        
        # 4. Renomear a coluna de Risco de Mercado para manter um padrão mais limpo
        df_fatores.rename(columns={'Rm_minus_Rf': 'Rm_Rf'}, inplace=True)
        
        print(f"\n✅ Matriz de Fatores limpa com sucesso! Total de dias: {len(df_fatores)}")
        print("\n--- Amostra dos Dados (Últimos 5 dias) ---")
        print(df_fatores.tail())
        
        # 5. Exportar a versão limpa para a LSTM consumir (Opcional, mas recomendado)
        caminho_saida = os.path.join(diretorio_dados, "fatores_nefin_limpos.csv")
        df_fatores.to_csv(caminho_saida)
        
        return df_fatores
        
    except FileNotFoundError:
        print(f"ERRO: O arquivo '{nome_arquivo}' não foi encontrado na pasta {diretorio_dados}.")
        print("Certifique-se de que o arquivo baixado foi salvo com esse nome na pasta data.")
        return None
    except Exception as e:
        print(f"Erro inesperado: {e}")
        return None

# ==========================================
# ÁREA DE EXECUÇÃO
# ==========================================
pasta_dados = r"C:\VSCodeWorkspace\TCC_Escrito\data"

# Certifique-se de que o nefin_factors.csv que você baixou está na pasta_dados!
df_fatores = processar_fatores_nefin_local(pasta_dados)
```
